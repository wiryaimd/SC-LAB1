```python
import numpy as np
from matplotlib import pyplot as plt
```

```python

# least regression
# y = b0 + b1.x
# y = b + mx 

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 5, 6, 5, 9, 10, 12, 14]

y = np.array(y)
A = np.vstack([x, np.ones(len(x))]).T
y = y[:, np.newaxis]


# alpha0 = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), y)
# print(alpha0)
alpha = np.dot(np.linalg.pinv(A), y)
print(alpha)
m = alpha[0]
b = alpha[1]

plt.plot(x, b + (m * x))
plt.plot(x, y, '.b')
plt.legend(["prediction", "x y"])
plt.xlabel("crit")
plt.show()

# print(A.T.T.T)

# np.info(np.ones)
# np.info(np.newaxis)
```
