```python
import numpy as np
```

```python
def f(x):
    return x**2+2
```

```python
#Mendefenisikan intergral dan titik a / b
a = 2
b = 5
n = 44
h = (b-a) / (n-1)

print(h)
```

```python
x = np.linspace(a,b,n)
print(x)
```

```python
y = f(x)
print(y)

y_left = y[ : n-1]
print(y_left)

y_right = y[1 : ]
print(y_right)

y_mid = (y_left + y_right) / 2
print(y_mid)
```

```python
#Hitung sum
left_rieman = h * np.sum(y_left)
right_rieman = h * np.sum(y_right)
middle_rieman = h * np.sum(y_mid)
```

```python
print(f'Left Rieman : {left_rieman}')
print(f'Right Rieman : {right_rieman}')
print(f'Middle Rieman : {middle_rieman}')
```

Trapzeoid

```python
trapezoid = (h/2) * (y_left[0] + y_right[-1] + (2 * np.sum(y[1:n-1])))
```

```python
print(f'Trapezoid : {trapezoid}')
```
