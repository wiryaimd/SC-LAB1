```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5,6,7,8]
y = [2,5,1,7,9,3,6,10]

x = np.array(x)
y = np.array(y)
```

```python
#Bentuk a yaitu hasil dari vstack dan transpose
A = np.vstack([x, np.ones(len(x))]).T
print(A)
```

```python
#Ubah y, dengan cara tambahin dimensinya
Y = y[ :, np.newaxis]
```

```python
alpha = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), Y)
#alpha[0] -> slope (m)
#alpha[1] -> intercept (b)
```

```python
#Since y = mx + b
#Parameter plt.plot(sumbu x, sumbu y, styling)
plt.plot(x,y,'bo')
plt.plot(x, alpha[0] * x + alpha[1], 'r')
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.legend(["x y coordinates", 'least square regression'])
plt.show()
```
